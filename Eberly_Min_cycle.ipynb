```python
def perp( a ) :
    b = empty_like(a)
    b[0] = -a[1]
    b[1] = a[0]
    return b

def find_start_node(nodes,node_coord):
    start_node = []
    minx = float('inf')
    for key,node in nodes.items():
        if node[0] <= minx:
            minx = node[0]
            start_node.append(key)
    if len(start_node) > 1:
        miny = float('inf')
        for i in range(len(start_node)):
            if nodes[i][1] < miny:
                miny = nodes[i][1]
            else:
                start_node.remove(i)
    return nodes[start_node[0]], node_coord[nodes[start_node[0]]]

def adj_nodes(start_key, edges):
    vnext = []
    for edge in edges:
        if edge[0] == start_key:
            vnext.append(edge[1])
    if len(vnext) == 0:
        pass
        #print "Vertex is end point."
    return vnext

def clockwise(nodes,vnext,start_key, v_prev):
    v_curr = np.asarray(nodes[start_key])
    if v_prev == None:
        v_prev = np.asarray([v_curr[0],(v_curr[1]-1)]) #This should be a vertical tangent to the start node at initialization.
    for v_adj in vnext[:]: #Again a copy so we can remove items
        d_curr = v_curr - v_prev
        d_next = np.asarray(nodes[v_adj]) - v_curr
        convex = np.dot(d_next,perp(d_curr)) #Both are convex and both are negative.
        d_adj = np.asarray(nodes[v_adj]) - v_curr
        if convex < 0:
            if (d_curr[0]*d_adj[1] - d_curr[1]*d_adj[0]) or (d_next[0]*d_adj[1]-d_next[1]*d_adj[0]) < 0:
                v_next = np.asarray(nodes[v_adj])
                d_next = d_adj 
                convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0]
        else:
            if (d_curr[0]*d_adj[1] - d_curr[1]*d_adj[0]) and (d_next[0]*d_adj[1]-d_next[1]*d_adj[0]) < 0:
                v_next = np.asarray(nodes[v_adj])
                d_next = d_adj 
                convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0] 
    prev_key = start_key
    return tuple(v_next.tolist()), node_coord[tuple(v_next.tolist())], prev_key

def counterclockwise(nodes, vnexts, start_key, prev_key):
    v_curr = np.asarray(nodes[start_key])
    v_prev = np.asarray(nodes[prev_key])
    for v_adj in vnexts: #Again a copy so we can remove items
        if v_adj == prev_key:
            continue
        d_curr = v_curr - v_prev
        d_next = np.asarray(nodes[v_adj]) - v_curr
        convex = np.dot(d_next,perp(d_curr))
        d_adj = np.asarray(nodes[v_adj]) - v_curr
        if convex < 0:
            if (d_curr[0]*d_adj[1] - d_curr[1]*d_adj[0]) and (d_next[0]*d_adj[1] - d_next[1]*d_adj[0]) > 0:
                v_next = np.asarray(nodes[v_adj])
                d_next = d_adj 
                convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0]
        else:
            if (d_curr[0]*d_adj[1] - d_curr[1]*d_adj[0]) or (d_next[0]*d_adj[1]-d_next[1]*d_adj[0]) > 0:
                v_next = np.asarray(nodes[v_adj])
                d_next = d_adj 
                convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0]
    prev_key = start_key
    try:
        return tuple(v_next.tolist()), node_coord[tuple(v_next.tolist())], prev_key
    except:
        return None, None, prev_key

def remove_edge(v0,v1,in_edges):
    def determine(v1,v0,edge):
        if v1 in edge and v0 in edge:
            print edge
            return False
        else:
            return True
    #in_edges[:] = [edge for edge in in_edges if v0 not in edge and v1 not in edge]
    edges[:] = [edge for edge in edges if determine(v1,v0,edge)]
    print "Out: ", in_edges
    return in_edges

def remove_heap(v0,sorted_nodes):
    sorted_nodes[:] = [x for x in sorted_nodes if x[0] != v0]
    return sorted_nodes
        
def remove_node(v0, nodes, nodes_coord):
    del nodes_coord[nodes[v0]]
    del nodes[v0]
    return nodes, nodes_coord
    
def extractisolated(nodes,node_coord,v0,primitives):
    primitives.append((v0))
    remove_node(v0, nodes, node_coord)
    return nodes, node_coord, primitives

def extractfilament(v0,v1, nodes, node_coord,sorted_nodes, ef_edges, primitives,cycle_edge, iscycle=False):
    for e in cycle_edge:
        if v0 and v1 in e:
            iscycle = True
    if iscycle == True:
    #This deletes edges that are part of a cycle, but does not add them as primitives.
        if len(adj_nodes(v0,ef_edges)) >= 3:
            ef_edges = remove_edge(v0,v1,ef_edges)
            v0 = v1
            if len(adj_nodes(v0, ef_edges)) == 1:
                v1 = adj_nodes(v0, ef_edges)[0]
        while len(adj_nodes(v0, ef_edges)) == 1:
            v1 = adj_nodes(v0, ef_edges)[0]
            #Here I need to do the cycle check again.
            iscycle = False
            for e in cycle_edge:
                if v0 and v1 in e:
                    iscycle = True
                    break
            if iscycle == True:
                ef_edges = remove_edge(v0,v1,ef_edges)
                nodes, node_coord = remove_node(v0, nodes, node_coord)
                sorted_nodes = remove_heap(v0, sorted_nodes)
                v0 = v1
            else:
                break
        if len(adj_nodes(v0, ef_edges)) == 0:
            nodes, node_coord = remove_node(v0, nodes, node_coord)
            sorted_nodes = remove_heap(v0, sorted_nodes)
    else:
        #Filament found
        primitive = []
        if len(adj_nodes(v0,edges)) >= 3:
            primitive.append(v0)
            ef_edges = remove_edge(v0,v1,ef_edges)
            v0 = v1
            if len(adj_nodes(v0, ef_edges)) == 1:
                v1 = adj_nodes(v0, ef_edges)[0]

        while len(adj_nodes(v0, ef_edges)) == 1:
            primitive.append(v0)
            v1 = adj_nodes(v0, ef_edges)[0]
            sorted_nodes = remove_heap(v0, sorted_nodes)
            ef_edges = remove_edge(v0, v1, ef_edges)
            print "Edges: ", edges
            nodes, node_coord = remove_node(v0, nodes, node_coord)
            v0 = v1
        
        primitive.append(v0)
        if len(adj_nodes(v0, edges)) == 0:
            sorted_nodes = remove_heap(v0, sorted_nodes)
            ef_edges = remove_edge(v0, v1, ef_edges)
            nodes, node_coord = remove_node(v0, nodes, node_coord)
        primitives.append((primitive))
        
    return sorted_nodes, ef_edges, nodes, node_coord, primitives

def extract_primitives(start_key,sorted_nodes, ep_edges, nodes, node_coord, primitives,minimal_cycles,cycle_edge):
    v0 = start_key
    visited = []
    sequence = []
    sequence.append(v0)

    #Find the CWise most vertex
    vnext = adj_nodes(start_key, ep_edges)
    start_node,v1,v_prev = clockwise(nodes,vnext,start_key,prev_key)
    v_curr = v1
    v_prev = v0
    
    #Find minimal cycle using CCWise rule
    process = True
    if v_curr == None:
        process = False
    elif v_curr == v0:
        process = False
    elif v_curr in visited:
        process = False
    
    while process == True:
        sequence.append(v_curr)
        visited.append(v_curr)
        vnext = adj_nodes(v_curr, ep_edges)
        v_curr_coords,v_next,v_prev = counterclockwise(nodes,vnext,v_curr, v_prev)
        v_curr = v_next
        if v_curr == None:
            process = False
        elif v_curr == v0:
            process = False
        elif v_curr in visited:
            process = False
            
    if v_curr is None:
        #Filament found, not necessarily at start_key
        sorted_nodes, ep_edges, nodes, node_coord, primitives = extractfilament(v0, adj_nodes(v0, ep_edges)[0],nodes, node_coord, sorted_nodes, ep_edges, primitives, cycle_edge)
    elif v_curr == v0:
        #Minimal cycle found
        primitive = []
        iscycle=True
        sequence.append(v0)
        minimal_cycles.append(tuple(sequence))
        #Remove the v0, v1 edges from the graph.
        edges = remove_edge(v0,v1,ep_edges)
        sorted_nodes = remove_heap(v0, sorted_nodes)#Not in pseudo-code, but in source.
        
        #Mark all the edges as being part of a minimal cycle.
        if len(adj_nodes(v0, ep_edges)) == 1:
            cycle_edge.append((v0, adj_nodes(v0, ep_edges)[0]))
            sorted_nodes, ep_edges, nodes, node_coord, primitives = extractfilament(v0, adj_nodes(v0, ep_edges)[0],nodes, node_coord, sorted_nodes, ep_edges, primitives,cycle_edge,  iscycle=True)
        if len(adj_nodes(v1, ep_edges)) == 1:
            cycle_edge.append((v1, adj_nodes(v1, ep_edges)[0]))
            sorted_nodes, ep_edges, nodes, node_coord, primitives = extractfilament(v1, adj_nodes(v1, ep_edges)[0],nodes, node_coord, sorted_nodes, ep_edges, primitives, cycle_edge, iscycle=True)
    
        for i,v in enumerate(sequence[1:-1]):
            cycle_edge.append((v,i+1))
            cycle_edge.append((i+1,v))
    else:
        #vcurr was visited earlier, so traverse the filament to find the end
        while len(adj_nodes(v0,ep_edges)) == 2:
            if adj_nodes(v0,ep_edges)[0] != v1:
                v1 = v0
                v0 = adj_nodes(v0,ep_edges)[0]
            else:
                v1 = v0
                v0 = adj_nodes(v0, ep_edges)[1]
        sorted_nodes, ep_edges, nodes, node_coord, primitives = extractfilament(v0,v1,nodes, node_coord, sorted_nodes, ep_edges, primitives,cycle_edge)
    return sorted_nodes, ep_edges, nodes, node_coord, primitives, minimal_cycles,cycle_edge
```
