#####Attributes for every point pattern

1. dist_to_node dict keyed by pointid with the value being a dict in the form {node: distance to node, node: distance to node}
2. obs_to_edge dict keyed by edge with the value being a dict in the form {pointID:(x-coord, y-coord), pointID:(x-coord, y-coord), ... }
3. obs_to_node
4. points geojson like representation of the point pattern.  Includes properties if read with attributes=True
5. snapped_coordinates dict keyed by pointid with the value being (x-coord, y-coord)

####Counts per edge are important, but should not be precomputed since we have different representations of the network (digitized and graph currently).  (Relatively) Uniform segmentation still needs to be done.

```python
counts = ntw.count_per_edge(ntw.pointpatterns['crimes'].obs_to_edge,
                            graph=False)
```

#####Moran's I using the digitized network

```python
#Binary Adjacency
ntw.contiguityweights(graph=False)
#Build the y vector
edges = ntw.w.neighbors.keys()
y = np.zeros(len(edges))
for i, e in enumerate(edges):
    if e in counts.keys():
        y[i] = counts[e]

#Moran's I
res = ps.esda.moran.Moran(y, ntw.w, permutations=99)
print dir(res)
```

####Moran's I using the graph representation to generate the W

Note that we have to regenerate the counts per edge, since the graph will have less edges.

```python
counts = ntw.count_per_edge(ntw.pointpatterns['crimes'].obs_to_edge, graph=True)

#Binary Adjacency
ntw.contiguityweights(graph=True)

#Build the y vector
edges = ntw.w.neighbors.keys()
y = np.zeros(len(edges))
for i, e in enumerate(edges):
    if e in counts.keys():
        y[i] = counts[e]

#Moran's I
res = ps.esda.moran.Moran(y, ntw.w, permutations=99)
print dir(res)
```

```python
n = 3 # number of points
lr = np.tile([0,0,1,1], n*(n-1)/2.)
rr = np.tile([0,1,0,1], n*(n-1)/2.)
dijleft = np.hstack([np.tile(np.hstack((np.repeat(p2n[i,0],2),np.repeat(p2n[i,1],2))),n-1-i) for i in range(n-1) ])
dijright = np.hstack([np.tile(p2n[i,:], 2*i) for i in range(1,n) ])
left = np.hstack([np.repeat(np.repeat([i],4),n-1-i) for i in range(n-1)])
right = np.hstack([np.repeat(np.repeat([i],4),i) for i in range(1,n)])

# dpn is nx2 ordered by points on row, column 1 is distance to its left node on its edge, column 2 is distance to right node
# d is the node to node shortest path distance matrix (precaluclated for the nework)
# four network distances are calculated for each pair of points
# d0 = dpn[i,0] + d[il,jl] + dpn[j,0]
# d1 = dpn[i,0] + d[il,jr] + dpn[j,1]
# d2 = dpn[i,1] + d[ir,jl] + dpn[j,0]
# d3 = dpn[i,1] + d[ir,jr] + dpn[j,1]
# and the minimum of the four is taken as the network distance between the pair
# this is vectorized so that res1 is of size (4*n*(n-1)/2), first four elements or d0,d1,d2,d3 for the first pair (0,1)
# next four are for pair (0,2) and so on until pair (n-2,n-1)
res1 = dpn[left,lr] + d[dijleft,dijright] + dpn[right,rr] 
sp_points = res1.reshape(n*(n-1)/2,4).min(axis=1)
```
