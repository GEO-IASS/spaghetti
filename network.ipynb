Once we have the regions the next step is to determine the pointers to the left and right regions for each edge. For this we need the x,y coords for each start and end node of the edge. We have to add these as attributes for each node.

```python
coords = (0,0), (-3,3), (0,5), (3,3), (3,-3), (0,-5), (-3,-3)
```

```python
for i in range(1,8):
    x,y = coords[i-1]
    G.node[i]['x'] = x
    G.node[i]['y'] = y
```

```python
G.nodes(data=True)
```

If nodes are ordered clockwise then the region is to the right of each of the directed edges. If nodes are ordered counterclockwise, then the region is to the left of each of the directed edges.

```python
pos = dict([ (i, coords[i-1]) for i in range(1,8) ]) 
```
